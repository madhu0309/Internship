**Create an array from 0 to 256 with step size 1 and data type is float and reshape it to 16 x 16 matrix**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr13 = np.arange(0,256,1,dtype=float)
arr13 = np.reshape(arr13,(16,16))
print(arr13.shape)
print(arr13)
plt.imshow(arr13,cmap = 'gray')
```
