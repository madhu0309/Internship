# DataFrames

```python
table1 = {'A':['a0','a1','a2','a3'],
         'B':['b0','b1','b2','b3'],
         'C':['c0','c1','c2','c3']}
```

```python
table2 = {'D':['d0','d1','d2','d3'],
         'E':['e0','e1','e2','e3'],
         'F':['f0','f1','f2','f3']}
```

```python
table3 = {'G':['g0','g1','g2','g3'],
         'H':['h0','h1','h2','h3'],
         'I':['i0','i1','i2','i3']}
```

```python
df1 = pd.DataFrame(table1,index=[0,1,2,3])
df2 = pd.DataFrame(table2,index = [4,5,6,7])
df3 = pd.DataFrame(table3,index = [8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

# Concatention

```python
# appending tables
df = pd.concat((df1,df2,df3))
df
```

```python
table1 = {'A':['a0','a1','a2','a3'],
         'B':['b0','b1','b2','b3'],
         'C':['c0','c1','c2','c3'],
         'key':['k0','k1','k2','k3']}

table2 = {'D':['d0','d1','d2','d3'],
         'E':['e0','e1','e2','e3'],
         'F':['f0','f1','f2','f3'],
         'key':['k1','k2','k0','k0']}
```

```python
left = pd.DataFrame(table1)
right = pd.DataFrame(table2)
```

```python
left
```

```python
right
```

# Merge

```python
inner = pd.merge(left,right,how='inner',on='key')
```
