```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Create an array from -10 to 10 with step of 1**

$plt.style.available$

```python
plt.style.use('fivethirtyeight')
```

```python
X = np.arange(-10,11,1,dtype=float)
y = X**2 
```

```python
plt.plot(X,y) # function
plt.xlabel('$X$') # x label
plt.ylabel('$X^2$') # y label
plt.title('$y = X^2$') # name your title
plt.show()
```

# Subplot

```python
# figsize = width , height
fig = plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.plot(X,X,'r') # function
plt.xlabel('$X$') # x label
plt.ylabel('$X$') # y label
plt.title('$y = X$') # name your title


plt.subplot(2,2,2)
plt.plot(X,X**2,'g')
plt.xlabel('$X$') # x label
plt.ylabel('$X^2$') # y label
plt.title('$y = X^2$') # name your title

plt.subplot(2,2,3)
plt.plot(X,X**3,'k:')
plt.xlabel('$X$') # x label
plt.ylabel('$X^3$') # y label
plt.title('$y = X^3$') # name your title

plt.subplot(2,2,4)
plt.plot(X,X**4,'--')
plt.xlabel('$X$') # x label
plt.ylabel('$X^4$') # y label
plt.title('$y = X^4$') # name your title
plt.show()
```

```python
fig.savefig('math.png')
```

# Plot within plot

```python
fig = plt.figure(figsize=(6,6))
axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) # bigger plot
axes2 = fig.add_axes([0.2,0.6,0.3,0.3]) # smaller plot
# x pos , y pos, width , height
axes1.plot(X ,X,'g')
axes1.set_xlabel('X')
axes1.set_ylabel('y = x')
axes2.plot(X,X**2,'m--')
axes2.set_xlabel('X')
axes2.set_ylabel('$X^2$')
plt.show()
```
