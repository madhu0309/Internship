```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Generate number from -10 to 10 using arange

plt.style.available

```python
plt.style.use('fivethirtyeight')
```

```python
X = np.arange(-10,11,1)
y = X**2 # y = x square
plt.plot(X,y)
plt.xlabel('X')
plt.ylabel('$X^2$')
plt.title('$y = x^2$')
plt.show()
```

# Subplot

```python
# width and height
fig = plt.figure(figsize=(15,15)) # adjusting figure size

plt.subplot(2,2,1) # area-1
plt.plot(X,X)
plt.xlabel('X')
plt.ylabel('$X$')
plt.title('$y = x$')

plt.subplot(2,2,2) # area-2
plt.plot(X,X**2)
plt.xlabel('X')
plt.ylabel('$X^2$')
plt.title('$y = x^2$')

plt.subplot(2,2,3) # area -3
plt.plot(X,X**3)
plt.xlabel('X')
plt.ylabel('$X^3$')
plt.title('$y = x^3$')

plt.subplot(2,2,4) # area -4
plt.plot(X,X**4)
plt.xlabel('X')
plt.ylabel('$X^4$')
plt.title('$y = x^4$')

plt.show()
```

```python
fig.savefig('srikanth.png')
```
