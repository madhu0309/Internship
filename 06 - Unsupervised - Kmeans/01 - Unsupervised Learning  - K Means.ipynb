```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Step -1 : Load dataset**

```python
# C:/Users/srikanth/Desktop/MLInternship/Batch1/day9
dataset = pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
X = dataset.iloc[:,[3,4]].values # independent varibles
```

**Step -2 :data visualization**

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], s= 100,edgecolors='black')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

**Step -3: Bulding machine leanring model**

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,15):

    kmean = KMeans(n_clusters=i)
    kmean.fit(X) # training 
    wcss.append(kmean.inertia_) # within cluster sum of square
```

**Elbow method**

```python
plt.figure(figsize=(10,6))
plt.plot(wcss)
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```

from above project it like to number of cluster is 4 but always good practice to take k+1 = 5 cluster

```python
model = KMeans(n_clusters=5)
```

```python
model.fit(X) # train model
y_means = model.predict(X)
```

```python
centroid = model.cluster_centers_
```

**Visualize data**

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],s= 100, c= y_means,edgecolors='black')
plt.scatter(centroid[:,0],centroid[:,1],color = 'red',s = 500,edgecolors='black')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
